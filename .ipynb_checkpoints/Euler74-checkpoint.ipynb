```python
from math import factorial

steps = [0 for i in range(3000001)]

steps[169] = steps[363601] = steps[1454] = 3
steps[871] = steps[45361] = 2
steps[872] = steps[45362] = 2


def sum_of_factorials(n):
    return sum(factorial(int(digit)) for digit in str(n))


def get_stepco(n):
    
    stepco = 1
    
    while 1:
        
        next = sum_of_factorials(n)
        
        if n == next:
            steps[n] = stepco
            return stepco
        
        if steps[next] > 0:
            stepco += steps[next]
            steps[n] = stepco
            return stepco

        return get_stepco(next) +1
        

co = 0
for n in range(1,1000000):
    
    if get_stepco(n) == 60:
        co += 1
    
print co
```

```python
import sys
import time
from math import factorial as fact

def fact_it(number):
    return sum([fact(int(n)) for n in str(number)])

if __name__ == "__main__":

    begin = time.time()

    final = list()

    history_dict = dict()

    for num in range(1,1000000):
        n = num
        chain = list()
        while n not in chain and n not in history_dict:
            chain.append(n)
            n = fact_it(n)

        chain_len = len(chain) if n not in history_dict else len(chain) + history_dict[n]
        history_dict[num] = chain_len
        if chain_len == 60:
            final.append(num)

    print("Total: ", len(final))

    print("Time: ", time.time()-begin)
```

```python
cache = {}

def get_chain(number, terms=None):
    if terms == None:
        terms = [number]
    if number in cache:
        res = cache[number]
    else:
        res = sum([factorial(int(x)) for x in list(str(number))])
        cache[number] = res
    if res in terms:   # We've hit a repeating number
        return terms
    terms.append(res)
    if get_chain(res, terms):
        return len(terms)

print sum([1 for i in xrange(1,1000000) if get_chain(i) == 60])
```
